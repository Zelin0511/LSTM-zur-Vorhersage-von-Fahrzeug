```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
file_path = 'pred_result_modified.csv'  # 请替换为实际文件路径
df = pd.read_csv(file_path)

# 提取重复列（假设默认情况下pandas会自动重命名重复列名）
# 确定两组列名
first_set = ['Gierwinkel', 'Geschw_X', 'Geschw_Y', 'Giergeschwindigkeit']
second_set = ['Gierwinkel.1', 'Geschw_X.1', 'Geschw_Y.1', 'Giergeschwindigkeit.1']

time = df['Time.1']

# 创建图表进行对比
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

for i, (col1, col2) in enumerate(zip(first_set, second_set)):
    row, col = divmod(i, 2)
    axs[row, col].plot(time, df[col1], label=col1, color='blue')
    axs[row, col].plot(time, df[col2], label=col2, color='red', linestyle='--')
    axs[row, col].set_xlabel('Time.1')
    axs[row, col].set_ylabel(col1)
    axs[row, col].legend()
    axs[row, col].set_title(f'Comparison of {col1} and {col2}')

plt.tight_layout()

# 保存图表为图片
output_path = 'comparison_plot.png'  # 保存在当前目录下
plt.savefig(output_path, dpi=300)

plt.show()
```
