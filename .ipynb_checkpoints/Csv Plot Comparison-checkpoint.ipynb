```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取csv文件
pred_df = pd.read_csv('pred_result.csv')
actual_df = pd.read_csv('actual.csv')

# 检查数据框前几行
print(pred_df.head())
print(actual_df.head())

# 如果时间列是以秒为单位的数值，确保是浮点或整数类型
pred_df['Time'] = pd.to_numeric(pred_df['Time'], errors='coerce')  # 转换为数值
actual_df['Time'] = pd.to_numeric(actual_df['Time'], errors='coerce')

# 排除无法解析的时间行
pred_df = pred_df.dropna(subset=['Time'])
actual_df = actual_df.dropna(subset=['Time'])

# 以Time为排序基准
pred_df = pred_df.sort_values(by='Time')
actual_df = actual_df.sort_values(by='Time')

# 合并数据框
merged_df = pd.merge(pred_df, actual_df, on='Time', suffixes=('_pred', '_actual'))

# 设置图片大小
plt.figure(figsize=(12, 8))

# 匹配的列名
columns = ['KS_X', 'KS_Y', 'Gierwinkel', 'Geschw_X', 'Geschw_Y', 'Giergeschwindigkeit']

# 依次作图
for col in columns:
    plt.figure()
    plt.plot(merged_df['Time'], merged_df[f'{col}_pred'], label=f'{col} (Prediction)')
    plt.plot(merged_df['Time'], merged_df[f'{col}_actual'], label=f'{col} (Actual)')
    plt.xlabel('Time (s)')  # 调整为秒
    plt.ylabel(col)
    plt.title(f'Comparison of {col} Over Time')
    plt.legend(loc='upper right')
    plt.grid(True)
    plt.savefig(f'{col}_comparison.png')  # 存为图片
    plt.show()
```
